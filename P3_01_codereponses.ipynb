# Question 3 : Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.

```python
df2['origine'] = 'animal'
df3['origine'] = 'vegétale'
```

```python
#je fais une concatenation du fichier animaux et vegetaux
df4=pd.concat([df2,df3])
```

```python
#J'utilise pivot table pour mieux synthétiser mes données et agréger les résultats
#Synthese sur les données du df4 pour obtenir les valeurs disponibles
pivot1=pd.pivot_table(df4, index=['Code zone','Zone','Code Produit','Produit','origine'] , columns='Élément',values='Valeur')
pivot1=pivot1.reset_index()
pivot1

```

```python
#Par la suite j'effectue une jointure entre mon pivot et mon csv population en selectionnant juste 2 données sur zone et valeur
df5=pd.merge(pivot1,df1[mask][['Zone','Valeur']] , on='Zone')
```

```python
#je vais calculer la disponibilité alimentaire en kcal/personne/jour  par produit par pays 
#je crée une nouvelle colonne 'Disponibilité almientaire(kcal)/ans' en convertissant
df5['Disponibilité almientaire(kcal)/ans']=df5['Disponibilité alimentaire (Kcal/personne/jour)']*365*df5['Valeur']*1000
```

```python
# je vais calculer la disponibilté alimentaire en kg protéine par ans par produit et par pays

df5['Disponibilité en protéines par kg']=(df5['Disponibilité de protéines en quantité (g/personne/jour)']/1000)*365*(df5['Valeur']*1000)
df5
```

# Question 4 : A partir de ces dernières informations, et à partir du poids de la disponibilité alimentaire (pour chaque pays et chaque produit), calculez pour chaque produit le ratio "énergie/poids", que vous donnerez en kcal/kg. 

```python
#Nous allons calculer le poid de la disponibilité alimentaire pour chaque pays et produit
df5['Disponibilité almientaire(kg)/ans']=df5['Disponibilité alimentaire en quantité (kg/personne/an)']*df5['Valeur']*1000
df5['Disponibilité almientaire(kg)/ans'].fillna(0, inplace=True)
```

```python
#Nous allons établire une divion entre disponibilité en kcal/ans et la disponibilité alimentaire en kg/ans
df5['Ratio energie poids en (kcal/kg)']=df5['Disponibilité almientaire(kcal)/ans']/df5['Disponibilité almientaire(kg)/ans']
df5['Ratio energie poids en (kcal/kg)'].fillna(0, inplace=True)
```

```python
df5['Ratio energie poids en (kcal/kg)'] = df5['Ratio energie poids en (kcal/kg)'].replace(np.inf, 0)
df5
```

# En suivant la même méthodologie, calculez également le pourcentage de protéines de chaque produit (pour chaque pays). Ce pourcentage est obtenu en calculant le ratio "poids de protéines/poids total" 

```python
#Nous allons établire une divion entre disponibilité en proteine par kg et la disponibilité alimentaire en kg/ans
#multiplier par 100 pour le %
df5['Ratio protéine/poids total en %']=df5['Disponibilité en protéines par kg']/df5['Disponibilité almientaire(kg)/ans']
df5['Ratio protéine/poids total en %'].fillna(0, inplace=True)
```

```python
df5['Ratio protéine/poids total en %'] = df5['Ratio protéine/poids total en %'].replace(np.inf, 0)
df5
```

# Question 10 : A partir des données téléchargées qui concernent la sous-nutrition, répondez à cette question : Quelle proportion de la population mondiale est considérée comme étant en sous-nutrition ?

```python
#importer le csv sousnutrition
df_sousnutri=pd.read_csv('fr_sousalimentation.csv')
#Afficher le type de valeurs
df_sousnutri.dtypes
```

```python
#Création d'une fonction qui me permettra aprés d'isolé l'année 2013
def format_année(Code_année):
    if Code_année == '2011-2013':
        return 2012

    elif Code_année == '2012-2014':
        return 2013
    
    elif Code_année == '2013-2015':
        return 2014
    
    elif Code_année == '2014-2016':
        return 2015
    
    elif Code_année == '2015-2017':
        return 2016
    
    elif Code_année == '2016-2018':
        return 2017
```

```python
#Application de la fonction grace à apply
df_sousnutri['Année'] = df_sousnutri['Année'].apply(format_année)
```

```python
df_sousnutri.fillna(0,inplace=True )
```

```python
#Aprés de multiples test à changer le type de Variable en float,j'ai pu voir que le problème
#était dans la ligne qui contenait Barbade
#Nous remplacons la la valeur <0.1 par 0 grace à la fonction.replace


df_sousnutri['Valeur']=df_sousnutri['Valeur'].replace('<0.1',0)
```

```python
#J'effectue ensuite un Cast avec la fonction to_numeric permet de convertir mon str en float
df_sousnutri['Valeur']=pd.to_numeric(df_sousnutri['Valeur'],downcast="float")
```

```python
df_sousnutri['is_undernour'] = False
df_sousnutri.loc[df_sousnutri['Valeur'] > 0, 'is_undernour'] = True
```

```python
#Je multiplie les données de la colonnes valeur pour les mettre en millions
df_sousnutri['Valeur']=df_sousnutri['Valeur']*1000000
```

```python
#DATAFRAME CREE A PARTIR DU CSV SOUS NUTRITION 
df_sousnutri
```

```python
#J'enleve La Chine du Dataframe pour ne pas fausser mes résultats
df_sousnutri = df_sousnutri[df_sousnutri.Zone != 'Chine']
```
